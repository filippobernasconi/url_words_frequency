```python
import urllib.request, urllib.error, urllib.parse, obo

url = 'https://www.corriere.it'

response = urllib.request.urlopen(url)
webContent = response.read()
webContent = str(webContent)
webContent = webContent.lower()

webContent
```

```python
import re

def removeHtmlTags(webContent):
    text = re.sub("\<(.*?)\>","",webContent)
    return text
    
webContent = removeHtmlTags(webContent)

def stripPunctuations(webContent):
    bad_chars = "?.,;:!"
    for char in webContent:
        if char in bad_chars:
            webContent = webContent.replace(char,"")
        
    return webContent
        
webContent = stripPunctuations(webContent)
print(webContent)

webContent = lower(webContent)
```

```python
text_list = webContent.split(" ")
text_list
```

```python
#tutte le parole nel testo, con ripetizioni

text_list_2 = []
for word in text_list:
    if len(word)> 4 and len(word)< 20 and "/" not in word and "\\" not in word and "(" not in word and "&" not in word:
        text_list_2.append(word)
        
print(text_list_2)
```

```python
#calcolo delle yes_words

list_yes_words = []
def saveYesWords(wordlist, yes_words):
    for word in wordlist:
        if word in yes_words:
            list_yes_words.append(word)
    return list_yes_words

#lista_di_prova = ["ciao","mamma","fashion","due","nonna","kfsf","tre"]

saveYesWords(text_list_2, yes_words)
list_yes_words

print("fase2")

def CountFrequency(text_list_2): 
      
    # Creating an empty dictionary  
    word_frequencies = {} 
    for items in text_list_2: 
        word_frequencies[items] = text_list_2.count(items)
    return word_frequencies
        
word_frequencies_yes_words = CountFrequency(list_yes_words)
word_frequencies_yes_words
```

```python
#calcolo stopwords


#elimina stopwords
list_saved_words = []
def avoidStopWords(wordlist, stopwords):
    for word in wordlist:
        if word not in stopwords:
            print(word)
            list_saved_words.append(word)
    return list_saved_words

list_saved_words = avoidStopWords(text_list_2, stopwords)
#list_saved_words
print("fase2")

#conta la frequenza delle parole senza stopwords

def CountFrequency(list_saved_words): 
      
    # Creating an empty dictionary  
    word_frequencies_saved_words = {} 
    for items in list_saved_words: 
        word_frequencies_saved_words[items] = list_saved_words.count(items)
    return word_frequencies_saved_words
        
word_frequencies_saved_words = CountFrequency(list_saved_words)
print(word_frequencies_saved_words)


#elimina le parole che hanno solo una occorrenza
print("fase3")

keyword_frequencies_dict = {}

for key, value in word_frequencies_saved_words.items():
    if value >= 2:
        keyword_frequencies_dict[key] = value

keyword_frequencies_dict
```

```python
#funziona, ma non c'è un metodo più semplice?

sorted_keyword_frequencies_dict = sorted(keyword_frequencies_dict.items(), key=lambda x: x[1], reverse=True)

sorted_keyword_frequencies_dict
```
